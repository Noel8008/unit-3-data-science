````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
titanic_df = pd.read_csv("titanic_passengers.csv")
````

## How many passengers were over the age of 35

````python
age = titanic_df["Age"]
age.plot.hist(edgecolor = 'black')
plt.title('Passenger Age')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
````

```
The graph clearly shows the age of how many or however passengers are the age between 1-80 
```

##   How many passengers' fare was over 30 dollars 


````python
fare = titanic_df['Fare']
fare.plot.hist(edgecolor = 'black')
plt.title('Passengers Fare')
plt.xlabel('Fare')
plt.ylabel('Numbers of Passengers')
````
