````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
titanic_df = pd.read_csv("titanic_passengers.csv")
````

## How many passengers were over the age of 35

````python
age = titanic_df["Age"]
age.plot.hist(edgecolor = 'black')
plt.title('Passenger Age')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
````

```
The graph clearly shows the age of how many or however passengers are the age between 1-80. Can see the range of people hitting each age and etc.
```

##   How many passengers' fare was over 30 dollars 


````python
fare = titanic_df['Fare']
plt.figure(figsize=(15, 6))
fare.plot.hist(edgecolor = 'black', bins=25)
plt.title('Passengers Fare')
plt.xlabel('Fare')
plt.ylabel('Numbers of Passengers')
````

```
Graph shows fare of passengers and how many passengers paid how much. Each passenger shows their fare range.
```

##   Did age affect your survival rates?

````python
titanic_df.plot.box(by='Survived', column='Age')
````

```
The green line in the box represents the average age of survivers and people who did die. The average age of people who died were actually 28 or 29ish and average age of people who survived were like 30'sish.
```

##   Did Pclass increase the fare?


````python
titanic_df.groupby(['Pclass'])['Fare'].mean().plot.bar()
````

```
The graph clearly shows that depending on your PClass your fare will increase. If you have a 1st class cabin your fare will drstically increase.
```

## Depending on peoples ages did they have more expensive fares and better pclass?



````python
titanic_df.plot.box(by='Pclass', column='Age')
titanic_df.plot.scatter('Fare', 'Age')
````

```
In the first graph it shows the average age of people in their designated Pclass. While the second graph shows the price for everyone and connects to their age.
```
